```python
import pandas as pd
```

```python
df_culture_news = pd.read_csv('culture_daum_news.csv')
df_society_news = pd.read_csv('society_daum_news.csv')
```

```python
df_culture_news
```

```python
df_society_news
del df_society_news['Unnamed: 0']
df_society_news
```

```python
df_news_total = pd.concat([df_culture_news, df_society_news])
df_news_total
```

```python
df_news_total.isna().sum()
```

```python
df_news_total = df_news_total.dropna(subset=['Content'])
df_news_total = df_news_total.drop_duplicates(subset=['Content'])
df_news_total
```

```python
df_news_total['MainCategory'].value_counts()
```

```python
df_news_total = df_news_total.reset_index(drop=True)
df_news_total
```

```python
df_news_total.to_csv('daum_news_team2_total.csv', index=False)
```
